# TASK #4: CREATE TESTING AND TRAINING DATASET & PERFORM DATA CLEANING

```python
X_cat=employee_df[['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus']]
```

```python
X_cat
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder=OneHotEncoder()
X_cat=onehotencoder.fit_transform(X_cat).toarray()
X_cat
```

```python
X_cat.shape
```

```python
X_cat=pd.DataFrame(X_cat)
```

```python
 X_cat
```

```python
# note that we dropped the target 'Atrittion'

X_numerical=employee_df[['Age','DailyRate','DistanceFromHome','Education','EnvironmentSatisfaction','RelationshipSatisfaction','StockOptionLevel','TotalWorkingYears','MonthlyIncome','MonthlyRate','NumCompaniesWorked','OverTime','PercentSalaryHike','PerformanceRating','TrainingTimesLastYear','WorkLifeBalance','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']]
X_numerical
```

```python
X_all=pd.concat([X_cat,X_numerical],axis=1)
```

```python
X_all
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X=scaler.fit_transform(X_all)
```

```python
X
```

```python
y=employee_df['Attritition']
y
```

# TASK #5: TRAIN AND EVALUATE A LOGISTIC REGRESSION CLASSIFIER

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
y_predict
```

```python
# Testing Set Performance
from sklearn.metrics import confusion_matrix,classification_report
print('Accuracy {} %'.format(100*accuracy_score(y_predict,y_test)))
```

```python
cm=confusion_matrix(y_predict,y_test)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_predict,y_test))
```

# TASK #6: TRAIN AND EVALUATE A RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=150)
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
# Testing Set Performance
print('Accuracy {} %'.format(100*accuracy_score(y_pred,y_test)))
```

```python
cm=confusion_matrix(y_pred,y_test)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_pred,y_test))
```
